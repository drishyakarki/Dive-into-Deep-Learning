```python
import torch
from d2l import torch as d2l
```

# Vanishing Gradients

```python
x = torch.arange(-10.0, 10.0, 0.2, requires_grad=True)
y = torch.sigmoid(x)
y.backward(torch.ones_like(x))

d2l.plot(x.detach().numpy(), [y.detach().numpy(), x.grad.numpy()], legend=['sigmoid', 'gradient'], figsize=(5, 3))
```

The sigmoid's gradient vanishes both when its inputs are large and when they are small.

# Exploding Gradient

```python
M = torch.normal(0, 1, size=(4, 4))
print('A single matrix \n', M)
for i in range(100):
    M = M @ torch.normal(0, 1, size=(4, 4))
print('After multiplying 100 matrices\n', M)
```
